1. Code below use to fix the broken encoding in csv before importing them to MySQL:

```python
import pandas as pd

#baca file dan perbaiki encoding csv
df = pd.read_csv("OnlineRetail.csv", dtype=str, encoding='latin1')

df = df.apply(lambda col: col.str.strip()if col.dtype == "object" else col)

df.to_csv("OnlineRetail.csv", index=False, encoding="utf-8")

print("Berhasil konversi ke UTF-8 bersih.")
```

2. Code below use to fix the writing of cleaned csv so Tableau can read them:

```python
import pandas as pd
import numpy as np

# 1. Tentukan nama kolom yang diinginkan
column_names = [
    'InvoiceNo', 'StockCode', 'Description', 'Quantity',
    'InvoiceDate', 'UnitPrice', 'CustomerID', 'Country', 'TotalSales'
]

try:
    # 2. Baca file asli
    # Memakai quoting=1 (QUOTE_ALL) karena data Anda dibungkus tanda kutip
    # na_values=['\\N'] akan mengubah \N menjadi nilai Null asli (NaN) di Python
    df = pd.read_csv(
        'OnlineRetail_cleaned.csv',
        names=column_names,
        header=None,
        quotechar='"',
        sep=',',
        na_values=['\\N'],
        dtype={'InvoiceNo': str, 'StockCode': str, 'CustomerID': str}, # Jaga ID tetap string agar tidak hilang angka nol di depan
        low_memory=False
    )

    # 3. Opsional: Pastikan InvoiceDate dalam format datetime agar Tableau langsung mengenalnya
    df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], errors='coerce')

    # 4. Simpan ke file baru yang "bersih" untuk Tableau
    # na_rep='' akan membuat nilai null menjadi kosong di CSV, yang dibaca sempurna oleh Tableau
    output_filename = 'OnlineRetail_Ready_for_Tableau.csv'
    df.to_csv(output_filename, index=False, quoting=1) # Menggunakan QUOTE_ALL agar lebih aman

    print(f"Berhasil! File disimpan sebagai: {output_filename}")
    print(f"Total baris yang diproses: {len(df)}")

except Exception as e:
    print(f"Terjadi kesalahan: {e}")
```
